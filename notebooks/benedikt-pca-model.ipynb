### Defining X and y

```python
y = data.iloc[:, -4] # y = "total_funding_usd"
X = data.iloc[:, 26:-4] #we've dropped first 25 columns b/c not numerical and we dropped last 3 columns for simplicity sake
X.shape
```

# Simple Data Preprocessing

### Changing dtypes

```python
# 1 column (articles) is still object type and thus has to be converted to int for the scaler to work
X['articles'] = X['articles'].str.replace(',', '')
X['articles'] = X['articles'].astype(float).astype(int)
```

### Checking NANs, imputing cloumns with few NANs, dropping columns with many NANs

```python
# list with all cloumns with NANs
my_NAN_dict = {k: v for k, v in dict(X.isna().sum()).items() if v > 0}
my_NAN_dict
```

```python
# Based on dict above, let's chose which columns to drop and which columns to impute
columns_to_drop = ['exit_year', 'closed_year', 'preseed_fund', 'seed_fund', 'series_a_fund', 'series_b_fund', 'series_c_fund', 'series_d_fund', 'series_e_fund', 'private_equity_fund']
columns_to_impute = ['last_funding_year', 'no_fund_rounds']
```

```python
# Dropping the columns specified above
X.drop(columns=columns_to_drop,inplace=True)
```

```python
# Imputing the columns specified above
imputer1 = SimpleImputer(strategy = "most_frequent")
imputer1.fit(X[['last_funding_year']])
X['last_funding_year'] = imputer1.transform(X[['last_funding_year']])

imputer2 = SimpleImputer(strategy = "most_frequent")
imputer2.fit(X[['no_fund_rounds']])
X['no_fund_rounds'] = imputer2.transform(X[['no_fund_rounds']])
```

```python
#Final check: make sure that everything above worked and that we really don't have any NANs left
{k: v for k, v in dict(X.isna().sum()).items() if v > 0}
```

### Train/validation/test split

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.20, random_state=1)
```

#### Scaling the features

```python
mm_scaler = MinMaxScaler()
mm_scaler.fit(X_train)

X_train_scaled = mm_scaler.transform(X_train)
X_test_scaled = mm_scaler.transform(X_test)
```
